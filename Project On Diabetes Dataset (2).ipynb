#### Calculate the mean of features for patients with diabetes and
#### Plot the mean values for each feature

```python
df_diabetes = diabetes_data[diabetes_data['Outcome'] == 1]
mean_features_diabetes = df_diabetes.mean()
plt.figure(figsize=(12, 6),dpi=100)
mean_features_diabetes.plot(kind='bar', color='salmon')
plt.title('Mean Features for Patients with Diabetes (Outcome == 1)')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Calculate the mean of features for patients without diabetes and
####  Plot the mean values for each feature

```python
df_diabetes_ = diabetes_data[diabetes_data['Outcome'] == 0]
mean_features_df_diabetes_ = df_diabetes_.mean()
plt.figure(figsize=(12, 6),dpi=100)
mean_features_df_diabetes_.plot(kind='bar', color='skyblue')
plt.title('Mean Features for Patients without Diabetes (Outcome == 0)')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
