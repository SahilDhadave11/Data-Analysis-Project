## Data Visualization

```python
sns.set()
plt.figure(figsize = (20,15))
sns.heatmap(diabetes_data.corr(), cmap='GnBu', annot=True)
plt.title('Correlation Graph')
plt.show()
```

#### The corelation between all the features are positive only. Hence, if ones values are increases corelated feature shows increament in the values.

#### Outcome is mostly affected by Glucose, Pregnancies, Age & BMI are greter than 0.2. Other factors like BloodPressure, DPF, Insulin & SkinThickness are below 0.2.

```python
sns.histplot(diabetes_data,x = diabetes_data['Pregnancies'],hue = 'Outcome')
```

#### Conclusion :-
                As pregnancies feature is categoriesed we can find the percentage of chances 
                of each categories to have diabetes.

```python
sns.histplot(diabetes_data,x = diabetes_data['Glucose'],hue = 'Outcome')
```

#### Conclusion :-
                probability of getting diabetes is higher when the glucose level is High. 
                (is above 150)

```python
sns.histplot(diabetes_data,x = diabetes_data['BloodPressure'],hue = 'Outcome')
```


#### Conclusion :-
                BloodPressure will not affect probabibly to have diabetes that much.

```python
sns.histplot(diabetes_data,x = diabetes_data['Insulin'],hue = 'Outcome')
```

#### Conclusion :-
                higher insulin levels tends to get diabetes.

```python
sns.histplot(diabetes_data,x = diabetes_data['BMI'],hue = 'Outcome')
```

#### Conclusion :-
     As we can observe here, BMI level from 25 to 35 gives lower chances of getting diabetes, 
     where as BMI level of above 40 has higher chances of getting diabetes.

```python
sns.histplot(diabetes_data,x = diabetes_data['SkinThickness'],hue = 'Outcome')
```

#### Conclusion :-
   Higher level of SkinThickness is associated with higher BMI and Insulin, which tends to increase the chances of getting diabetes. But direct effect of SkinThickness can be negligible.

```python
sns.histplot(diabetes_data,x = diabetes_data['DiabetesPedigreeFunction'],hue = 'Outcome')
```

#### Conclusion :-
            DPF has not that much effect on outcome.

```python
sns.histplot(diabetes_data,x = diabetes_data['Age'],hue = 'Outcome')
```

#### Conclusion :-
Higher the mean age higher the chance of getting diabetes. above 40 age it has much more tendency to get diabites.Now lets analyze main features that affect Outcome the most.

```python
# Plotting the heatmap to check the correlation between the Target Label and other features
sns.heatmap(diabetes_data.corr()[['Outcome']].sort_values(by='Outcome', ascending=False), vmin=-1, vmax=1,
            annot=True, cmap='GnBu')
```

```python
plt.style.use('seaborn')
fig, data = plt.subplots(figsize=(10, 8))
scatter = data.scatter(diabetes_data.index, diabetes_data['Pregnancies'], c=diabetes_data['Outcome'], cmap='RdYlGn', s=40)
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='g', markersize=10, label='Not Affected'),
                   plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='r', markersize=10, label='Affected')]
data.legend(handles=legend_elements, title='Outcome')
data.set_title('Scatter Plot of Glucose vs Outcome')
plt.show()
```

* But with higher BMI and lower Pregnancies will give the higher chances of diabetes and vice versa.


* Hence we can conclude that BMI can be higher deciding factor than Pregnancies.

* With Higher Glucose values above 150 and lower BMI will also get higher chances of Diabetes, 
and vice versa.
